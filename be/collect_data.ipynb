```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader as web
import yfinance as yfin

from pandas_datareader import data as pdr
```

```python
list_company = [
    'MMM',
    'AXP',
    'AAPL',
    'BA',
    'CAT',
    'CVX',
    'CSCO',
    'KO',
    'DIS',
    'DD',
    'XOM',
    'GE',
    'GS',
    'HD',
    'IBM',
    'INTC',
    'JNJ',
    'JPM',
    'MCD',
    'MRK',
    'MSFT',
    'NKE',
    'PFE',
    'PG',
    'TRV',
    'RTX',
    'UNH',
    'VZ',
    'V',
    'WMT'
]
```

```python
# declare variable
train_start_date = '2012-01-01'
train_end_date = '2021-01-01'
test_start_date = '2021-01-01'
test_end_date = '2023-01-01'
yfin.pdr_override()
```

```python
for company in list_company:
  data = pdr.get_data_yahoo(company, start=train_start_date, end=train_end_date)
  df = pd.DataFrame(data)
  df.to_csv(f"train_data/{company.lower()}.csv")
```

```python
for company in list_company:
  data = pdr.get_data_yahoo(company, start=test_start_date, end=test_end_date)
  df = pd.DataFrame(data)
  df.to_csv(f"test_data/{company.lower()}.csv")
```
